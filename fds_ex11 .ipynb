```python
import matplotlib.pyplot as plt
```

```python
days = list(range(1, 11))  # Days of the month
sales = [100, 120, 90, 150, 130, 170, 160, 180, 200, 220]
```

```python
plt.plot(days, sales, marker='o', color='blue')
plt.title("Sales Over Time (Line Plot)")
plt.xlabel("Day of Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
plt.scatter(days, sales, color='green')
plt.title("Sales Over Time (Scatter Plot)")
plt.xlabel("Day of Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
plt.bar(days, sales, color='purple')
plt.title("Sales Over Time (Bar Plot)")
plt.xlabel("Day of Month")
plt.ylabel("Sales")
plt.show()
```
