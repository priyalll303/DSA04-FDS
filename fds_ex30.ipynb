```python
from sklearn.tree import DecisionTreeRegressor, export_text
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\DELL\Downloads\fds_csv files\used_cars.csv")
X = df.drop("price", axis=1)
y = df["price"]
```

```python
model = DecisionTreeRegressor()
model.fit(X, y)
```

```python
features = [float(input(f"Enter {col}: ")) for col in X.columns]
pred_price = model.predict([features])
print(f"Predicted Price: ₹{pred_price[0]:,.2f}")
```

```python

tree_rules = export_text(model, feature_names=list(X.columns))
print("\nDecision Path:\n", tree_rules)
```
