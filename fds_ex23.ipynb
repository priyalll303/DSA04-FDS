```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
df = pd.read_csv(r"C:\Users\DELL\Downloads\clinical_trial.csv")
control = df[df['group'] == 'control']['outcome']
treatment = df[df['group'] == 'treatment']['outcome']
```

```python
t_stat, p_val = ttest_ind(control, treatment)
```

```python
plt.boxplot([control, treatment], tick_labels=['Control', 'Treatment'])
plt.title(f'T-test p-value: {p_val:.4f}')
plt.ylabel("Outcome Measure")
plt.show()
```
